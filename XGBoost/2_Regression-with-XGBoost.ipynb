### Common Regression Metrics

- RMSE: Root Mean Squared Error
- MAE: Mean Absolute Error

#### RMSE

Computing RMSE
- taking the difference between the actual and predicted (ie. the error)
- squaring the error
- Computing their mean
- and taking the Square Root of the mean

This allows us to treat positive and negative values equally.

#### MAE

Sums the absolute error differences.

Common Regresion Algorithms:
- Linear Regression
- Decision Trees
