### Regression Mechanics

Regression fit a line through a dataset, for the best fit line.

y=ax+b
- y = Target
- x = single feature
- a,b = parameters of the model

### How do we choose a, b?

- Best to define an error function for any given line and choose the line that maximizes that error.
- This essentially means we want the line that best fits the data, or in other words that minimizes the distance between every point and the line, ie the **residual distance** between the point and the line

This is called the **Loss Function**

- Positive and negative residual distances from the line cancel each other out.
- This is why we use the (minimize) the **sum of the squares of the residuals**.

### OLS: Ordinary Least Squares

- THIS is the minimizing sum of squares of residuals
